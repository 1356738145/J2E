# XSLX转JSON格式

```python
import pandas as pd
import json


def excel_to_json(file_path):
    df = pd.read_excel(file_path)
    result = {}
    current_first_level = None
    current_second_level = None
    for index, row in df.iterrows():
        first_level = row.iloc[0]
        second_level = row.iloc[1]
        third_level = row.iloc[2]
        # 如果一级标签不为空，更新当前一级标签并在结果中创建相应的键
        if pd.notna(first_level):
            current_first_level = first_level
            result[current_first_level] = {}
        # 如果二级标签不为空，更新当前二级标签并在当前一级标签下创建相应的键
        if pd.notna(second_level):
            current_second_level = second_level
            result[current_first_level][current_second_level] = []
        # 如果三级标签不为空，将其添加到当前二级标签的列表中
        if pd.notna(third_level):
            result[current_first_level][current_second_level].append(third_level)
    return result


def main():
    file_path = "左点非满问题类别标签整理.xlsx"  # 请将此路径替换为实际的 Excel 文件路径
    result_dict = excel_to_json(file_path)
    # 将字典转换为 JSON 格式，添加 ensure_ascii=False 参数
    json_result = json.dumps(result_dict, indent=4, ensure_ascii=False)
    print(json_result)


if __name__ == "__main__":
    main()
```

# 对话四级标签打标

```python
import pandas as pd
import json
import re

# 假设数据存储在名为 'data.csv' 的文件中
file_path = 'data (21).csv'
df = pd.read_csv(file_path)

# 定义函数来解析JSON数据并提取键值对
def extract_json(json_str):
    try:
        data = json.loads(json_str)
        labels = data.get('一级标签', None)
        sub_labels = data.get('二级标签', None)
        sub_sub_labels = data.get('三级标签', None)
        sub_sub_sub_labels = data.get('四级标签', None)
        reason = data.get('具体原因', None)
        return labels, sub_labels, sub_sub_labels, sub_sub_sub_labels, reason
    except json.JSONDecodeError:
        return None, None, None, None, None

# 应用函数并创建新的列
df[['一级标签', '二级标签', '三级标签', '四级标签', '具体原因']] = df['deepseek-r1'].apply(lambda x: pd.Series(extract_json(x)))

# 定义函数来清理文本列中的非法字符
def clean_text(text):
    if isinstance(text, str):
        # 移除不能在 Excel 中使用的控制字符
        cleaned = re.sub(r'[\x00-\x1F\x7F]', '', text)
        return cleaned
    return text

# 需要清理的列
columns_to_clean = ['contents', 'industry', 'labelList', '一级标签', '二级标签', '三级标签', '四级标签', '具体原因', 'deepseek-r1推理过程']
for col in columns_to_clean:
    df[col] = df[col].apply(clean_text)

# 删除包含空字符串或全为空格的行
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df = df[(df[columns_to_clean] != '').all(axis=1)]

# 保存处理后的数据为 Excel 文件
output_excel_path = 'tb_output_data.xlsx'
df[['contents', 'industry', 'labelList', '一级标签', '二级标签', '三级标签', '四级标签', '具体原因', 'deepseek-r1推理过程']].to_excel(output_excel_path, index=False)
print(f"数据已保存到 {output_excel_path}")
```

# 责任占比分析


```python
import pandas as pd
import json

# 读取 CSV 文件
file_path = 'tb.csv'
df = pd.read_csv(file_path)

# 定义要匹配的一级场景表头
primary_scenes = ['客服责任', '店铺责任', '物流责任']

# 定义函数来解析 JSON 数据并提取键值对
def extract_json(json_str):
    try:
        # 处理双引号问题
        data = json.loads(json_str.replace('""', '"'))
        # 提取业务问题
        business_issue = data.get('业务问题', None)
        primary_scene_list = data.get('一级场景', [])
        secondary_scene_list = data.get('二级场景', [])

        result = {
            '业务问题': business_issue
        }

        # 初始化一级场景列的值为 None
        for scene in primary_scenes:
            result[scene] = None

        # 处理一级场景和二级场景
        secondary_index = 0
        for primary_scene in primary_scene_list:
            primary_label = primary_scene.strip()
            if primary_label in primary_scenes:
                sub_scenes = []
                # 遍历二级场景，直到遍历完或者一级场景对应完
                while secondary_index < len(secondary_scene_list):
                    sub_scene = secondary_scene_list[secondary_index]
                    sub_scenes.append(sub_scene)
                    secondary_index += 1
                    # 如果一级场景只有一个，继续添加二级场景
                    if len(primary_scene_list) == 1 and secondary_index < len(secondary_scene_list):
                        continue
                    else:
                        break
                result[primary_label] = ', '.join(sub_scenes)

        return pd.Series(result)
    except json.JSONDecodeError:
        result = {
            '业务问题': None
        }
        for scene in primary_scenes:
            result[scene] = None
        return pd.Series(result)

# 应用函数并创建新的列
new_df = df['deepseek-r1'].apply(extract_json)

# 合并原始数据和新列
final_df = pd.concat([df, new_df], axis=1)

# 保存处理后的数据为 Excel 文件
output_excel_path = 'tb.xlsx'
final_df[['contents', 'industry', 'labelList', '一级标签',  '二级标签', '具体原因',  'deepseek-r1推理过程', 'deepseek-r1', '客服责任', '店铺责任', '物流责任', '业务问题']].to_excel(output_excel_path, index=False)
print(f"数据已保存到 {output_excel_path}")
```

```python
import pandas as pd
import json

# 读取 CSV 文件
file_path = 'pdd.csv'  # 请将此替换为你的 CSV 文件实际路径
df = pd.read_csv(file_path)

# 定义函数来解析 JSON 数据并提取业务问题
def extract_business_issue(json_str):
    try:
        # 处理双引号问题
        json_str = json_str.replace('""', '"')
        data = json.loads(json_str)
        return data.get('业务问题', None)
    except (json.JSONDecodeError, AttributeError):
        return None

# 应用函数并创建新的列
df['业务问题'] = df['deepseek-r1'].apply(extract_business_issue)

# 保存处理后的数据为新的 Excel 文件
output_file_path = 'pdd_final.xlsx'
df.to_excel(output_file_path, index=False)
print(f"数据已保存到 {output_file_path}")
```
